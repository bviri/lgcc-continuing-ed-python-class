```python
select_school_1 = schooldata.loc[['84X539']]
# this is the first code i tried using to filter out schooldata. not sure why it wasn't working, despite following the correct syntax and copypasting the names of the schooles
# the error said "None of [Index(['84X539'], dtype='object')] are in the [index]"
```

```python
select_school_1 = schooldata.loc[schooldata['School DBN'] == '84X539']
# this is the code i got to work to filter a row. this filters out the rows with the district number 84X539
select_school_1.head(50)
```

```python
newvisions = select_school_1[['School Name','Year','Total Tested','Regents Exam','Number Scoring 65 or Above']]
newvisions.head(10)
# separating data for new visions charter school for those passing
```

```python
newvisions.drop('School Name', axis=1, inplace = True)
# dropping the school name because i've only selected one school
newvisions.groupby(['Regents Exam'])['Number Scoring 65 or Above'].count().sort_values(ascending=False).plot(kind="bar",title="Number of Testees that Passed a Regents Exam Per Subject",ylabel="Number of Testees Scoring 65 or Above",color="purple")
```

```python
# awesome! one bar graph worked. i was able to check one school's passing students. however, i want to compare multiple schools with students who pass (score above 65), not just one school.
# so i have to filter and group differently

schooldata.loc[['University Neighborhood High School','Marta Valle High School','Bard High School Early College','The Urban Assembly School for Emergency Management']].groupby(['School Name'])['Number Scoring 65 or Above'].count()
# at first tried to separate by strings. however, kept getting an error that they were missing?
```

```python
# first let's drop the values that are null
diabetesdata.isnull().sum()
diabetesdata.dropna()
```

```python
datadropped = ['encounter_id', 'weight', 'payer_code', 'insulin', 'metformin','repaglinide','nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide', 'rosiglitazone', 'acarbose', 'troglitazone', 'miglitol', 'tolazamide', 'examide', 'citoglipton', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone']
diabetesdata.drop(datadropped, inplace= True, axis=1)

# i won't be measuring medications in my data, and the weight column, while it may have been useful, seems to only have ?
# so i'm going to drop these columns and others i find unnecessary
```

```python
diabetesdata.head()
```

```python
diabetesfemale = diabetesdata[diabetesdata['gender']== 'Female']
diabetesfemale.groupby(['race'])['patient_nbr'].count()
```

```python
# the ? identifies people who didn't list their race, or their race is unknown
diabetesfemale["race"] = diabetesfemale["race"].replace("?", "Unknown")
```

```python
diabetesfemale.groupby(['race'])['patient_nbr'].count()
```

```python
# but... wait, race is one of the components i'm working with. the unknown value causes complications, so let's drop it
diabetesfemale = diabetesfemale[diabetesfemale["race"].str.contains("Unknown") == False]
```

```python
diabetesfemale.groupby(['race'])['patient_nbr'].count()
```

```python
sns.boxplot(x='race', y='number_diagnoses', data=diabetesfemale, showfliers=True, palette="husl").set_title("Number of Diagnoses by Female Patients' Race")
plt.xticks(rotation=30)

# the box plot is great at showing the medians and the max/mins, however, due to the sample sizes i feel like this data can be misconstrued
# it is interesting to note that it seems there are higher medians and percentiles of other races, but that white women had more outliers
```
